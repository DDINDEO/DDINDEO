```python
import pandas as pd
import numpy
from numpy import set_printoptions

# --- Lab 2: ë°ì´í„° ì „ì²˜ë¦¬ (í‘œì¤€í™”) ---
from sklearn.preprocessing import StandardScaler

# --- Lab 3: íŠ¹ì„± ì¤‘ìš”ë„ (Feature Importance) ---
from sklearn.ensemble import ExtraTreesClassifier     #

# --- 1. ë°ì´í„° ë¡œë”© (ìƒˆ íŒŒì¼) ---
filename = 'givemesomecreidt.csv'
# ğŸ’¡ 78281ë²ˆ ì¤„ì˜ ì˜¤ë¥˜ë¥¼ 'ë¬´ì‹œ(skip)'í•˜ëŠ” ì˜µì…˜ ì¶”ê°€
data = pd.read_csv(filename, on_bad_lines='skip')

# --- 2. ì´ìƒì¹˜ ì²˜ë¦¬ (Outlier Capping) â­ ---
# (PPTì˜ ë…¼ë¦¬ë¥¼ ë§ì¶”ê¸° ìœ„í•´ ì´ìƒì¹˜ ì²˜ë¦¬)
outlier_cols = ['NumberOfTime30-59DaysPastDueNotWorse',
                'CombinedDefaulted'] # (ìƒˆ ë³€ìˆ˜ ì´ë¦„)

for col in outlier_cols:
    if col in data.columns:
        # ê°’ì´ 10ì„ ì´ˆê³¼í•˜ë©´ 10ìœ¼ë¡œ ê³ ì •
        data[col] = data[col].apply(lambda x: 10 if x > 10 else x)

# --- 3. X (í”¼ì²˜ 8ê°œ), Y (íƒ€ê²Ÿ 1ê°œ) ë¶„ë¦¬ ---
Y = data['SeriousDlqin2yrs'].values
X_df = data.drop('SeriousDlqin2yrs', axis=1)
X = X_df.values
feature_names = X_df.columns.tolist() # 8ê°œ í”¼ì²˜ ì´ë¦„

# --- 4. ë°ì´í„° ìŠ¤ì¼€ì¼ë§ (Lab 2) ---
# (10ê°œ -> 8ê°œ í”¼ì²˜ ì „ì²´ë¥¼ í‘œì¤€í™”)
scaler = StandardScaler().fit(X)
standardizedX = scaler.transform(X)

# --- 5. íŠ¹ì„± ì¤‘ìš”ë„ (Feature Importance) ì‹¤í–‰ (Lab 3) ---
print("--- 1-4. íŠ¹ì„± ì¤‘ìš”ë„ (ExtraTrees) ê²°ê³¼ (8ê°œ í”¼ì²˜) ---")


model = ExtraTreesClassifier(n_estimators=100, random_state=7)
# (!!) í‘œì¤€í™”ëœ 'standardizedX' (8ê°œ í”¼ì²˜)ë¡œ í›ˆë ¨
model.fit(standardizedX, Y)

# (PPTë¥¼ ìœ„í•´ ì¤‘ìš”ë„ë¥¼ í”¼ì²˜ ì´ë¦„ê³¼ í•¨ê»˜ ë³´ê¸° ì‰½ê²Œ ì¶œë ¥)
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': model.feature_importances_})
print(importance_df.sort_values(by='Importance', ascending=False))
```
